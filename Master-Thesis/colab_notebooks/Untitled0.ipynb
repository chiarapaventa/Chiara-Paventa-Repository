```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

## Data

```python
path = '/content/drive/MyDrive/Tesi/data/Dati/'
results_path = path + 'results/'
```

```python
results_load_SA = pd.read_csv(results_path + 'load_test_predictions_ARIMA_SARIMA.csv', delimiter=',', index_col='datetime')
results_load_LG = pd.read_csv(results_path + 'load_test_predictions_GRU.csv', delimiter=',', index_col='datetime')
results_load_w_SA = pd.read_csv(results_path + 'load_weather_test_predictions_ARIMA_SARIMA.csv', delimiter=',', index_col='datetime')
results_load_w_LG = pd.read_csv(results_path + 'load_weather_test_predictions_LSTM_GRU.csv', delimiter=',', index_col='datetime')

results_prod_SA = pd.read_csv(results_path + 'prod_test_predictions_ARIMA_SARIMA.csv', delimiter=',', index_col='datetime')
results_prod_L = pd.read_csv(results_path + 'prod_test_predictions_LSTM.csv', delimiter=',', index_col='datetime')
results_prod_G = pd.read_csv(results_path + 'prod_test_predictions_GRU.csv', delimiter=',', index_col='datetime')
results_prod_w_SA = pd.read_excel(results_path + 'prod_weather_test_predictions_ARIMA_SARIMA.xlsx', index_col='datetime')
results_prod_w_LG = pd.read_csv(results_path + 'prod_weather_test_predictions_LSTM_GRU.csv', delimiter=',', index_col='datetime')

results_load_SA.drop(results_load_SA.columns[0], axis=1, inplace=True)
results_prod_SA.drop(results_prod_SA.columns[0], axis=1, inplace=True)
results_load_w_SA.drop(results_load_w_SA.columns[0], axis=1, inplace=True)
results_load_w_LG.drop(results_load_w_LG.columns[0], axis=1, inplace=True)
results_prod_w_LG.drop(results_prod_w_LG.columns[0], axis=1, inplace=True)

column_to_move = results_prod_w_SA['produzionePV']
results_prod_SA.insert(0, 'produzionePV', column_to_move)
```

# Load

```python
mse_values = []
rmse_values = []
mae_values = []

for index, row in results_load_SA.iterrows():
    mse_value = mean_squared_error([row['carico']], [row['arima']])
    rmse_value = np.sqrt(mse_value)
    mae_value = mean_absolute_error([row['carico']], [row['arima']])

    mse_values.append(mse_value)
    rmse_values.append(rmse_value)
    mae_values.append(mae_value)

results_load_SA['MSE_ARIMA'] = mse_values
results_load_SA['RMSE_ARIMA'] = rmse_values
results_load_SA['MAE_ARIMA'] = mae_values
```
