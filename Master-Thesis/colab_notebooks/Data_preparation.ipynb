## Data Load

```python
data_load = data_inputs[['carico']]
```

```python
# Creating timestamps
timestamps = pd.date_range(start="2013-01-01", periods=17520, freq="30min")
data_load.insert(0, 'date', timestamps)

# Renaming first column
data_load = data_load.rename(columns={'date': 'datetime'})

# Resample the data to hourly intervals by taking the mean of every two 30-minute intervals
data_load = data_load.resample('H', on='datetime').mean()

data_load = data_load.reset_index()
data_load['datetime'] = pd.to_datetime(data_load['datetime'])
data_load.dtypes
```

##Data Load with One-Hot Encoder

```python
#data_load_weather = data_load_weather.reset_index()
data_load_weather['datetime'] = pd.to_datetime(data_load_weather['datetime'])
data_load_weather.dtypes
#data_load_weather = data_load_weather.set_index("datetime")
```

```python
# One hot encoder
data_load_weather['year'] = data_load_weather['datetime'].dt.year
data_load_weather['month'] = data_load_weather['datetime'].dt.month
data_load_weather['day'] = data_load_weather['datetime'].dt.day
data_load_weather['hour'] = data_load_weather['datetime'].dt.hour

columns_to_encode = ['month']
data_to_encode = data_load_weather[columns_to_encode]

# Perform one-hot encoding
encoder = OneHotEncoder()
encoded_data = encoder.fit_transform(data_to_encode)

# Create a new DataFrame with encoded columns
encoded_df = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out(columns_to_encode))

# Concatenate the encoded DataFrame with the original DataFrame
df_load_encoded = pd.concat([data_load_weather, encoded_df], axis=1)
```

```python
# Save on csv file
df_load_encoded.to_excel("data_load_weather_encoded.xlsx", index=True)
```
